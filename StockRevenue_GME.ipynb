```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {"User-Agent": "Mozilla/5.0"}
html = requests.get(url, headers=headers).text
soup = BeautifulSoup(html, "html.parser")

# Find all tables
tables = soup.find_all("table")
print("Number of tables found:", len(tables))

# Parse revenue table if available
if len(tables) >= 2:
    tesla_revenue = pd.read_html(str(tables[1]))[0]
    tesla_revenue.columns = ['Date', 'Revenue']
    tesla_revenue.dropna(inplace=True)

    tesla_revenue["Revenue"] = tesla_revenue["Revenue"].replace(r"\$", "", regex=True)
    tesla_revenue["Revenue"] = tesla_revenue["Revenue"].replace(",", "", regex=True)
    tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]

    print(tesla_revenue.tail())
else:
    print("Revenue table not found on the page.")
```

```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objs as go

# Step 1: Get Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)

# Step 2: Scrape Tesla revenue data from MacroTrends
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Step 3: Find table containing "Quarterly Revenue"
tables = soup.find_all("table")
tesla_revenue = None
for table in tables:
    if "Tesla Quarterly Revenue" in table.text:
        tesla_revenue = pd.read_html(str(table))[0]
        break

if tesla_revenue is None:
    raise Exception("Tesla revenue table not found.")

# Step 4: Clean the revenue data
tesla_revenue.columns = ['Date', 'Revenue']
tesla_revenue['Revenue'] = tesla_revenue['Revenue'].replace('[\$,]', '', regex=True)
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != '']

# Step 5: Define graph function
def make_graph(stock_data, revenue_data, stock_name):
    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=stock_data['Date'],
        y=stock_data['Close'],
        name="Stock Price",
        line=dict(color='blue')
    ))

    fig.add_trace(go.Scatter(
        x=revenue_data['Date'],
        y=revenue_data['Revenue'].astype(float),
        name="Revenue",
        line=dict(color='green'),
        yaxis='y2'
    ))

    fig.update_layout(
        title=f"{stock_name} Stock Price vs Revenue",
        xaxis=dict(title='Date'),
        yaxis=dict(title='Stock Price (USD)', side='left'),
        yaxis2=dict(title='Revenue (Millions USD)', overlaying='y', side='right'),
        legend=dict(x=0.1, y=1.1, orientation="h"),
        xaxis_rangeslider_visible=True,
        template='plotly_dark'
    )

    fig.show()

# Step 6: Call function
make_graph(tesla_data, tesla_revenue, 'Tesla')
```

```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objs as go

# Step 1: Download GME stock data
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)

# Step 2: Scrape GME Quarterly Revenue from MacroTrends
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

tables = soup.find_all("table")
gme_revenue = None
for table in tables:
    if "GameStop Quarterly Revenue" in table.text:
        gme_revenue = pd.read_html(str(table))[0]
        break

if gme_revenue is None:
    raise Exception("GameStop revenue table not found.")

# Step 3: Clean up the DataFrame
gme_revenue.columns = ['Date', 'Revenue']
gme_revenue['Revenue'] = gme_revenue['Revenue'].replace('[\$,]', '', regex=True)
gme_revenue.dropna(inplace=True)
gme_revenue = gme_revenue[gme_revenue['Revenue'] != '']

# Step 4: Define plot function (same as before)
def make_graph(stock_data, revenue_data, stock_name):
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=stock_data['Date'],
        y=stock_data['Close'],
        name="Stock Price",
        line=dict(color='blue')
    ))
    fig.add_trace(go.Scatter(
        x=revenue_data['Date'],
        y=revenue_data['Revenue'].astype(float),
        name="Revenue",
        line=dict(color='green'),
        yaxis='y2'
    ))
    fig.update_layout(
        title=f"{stock_name} Stock Price vs Revenue",
        xaxis=dict(title='Date'),
        yaxis=dict(title='Stock Price (USD)', side='left'),
        yaxis2=dict(title='Revenue (Millions USD)', overlaying='y', side='right'),
        legend=dict(x=0.1, y=1.1, orientation="h"),
        xaxis_rangeslider_visible=True,
        template='plotly_dark'
    )
    fig.show()

# Step 5: Plot the dashboard for GameStop
make_graph(gme_data, gme_revenue, 'GameStop')
```
